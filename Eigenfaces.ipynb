# Eigenfaces

```python
def load_img(path: str):
    img = Image.open(path).convert('L')
    return np.array(img.getdata(), dtype=np.uint8)
```

```python
def arr_to_img(arr):
    return Image.fromarray(np.reshape(arr, (250, 250))).convert('L')
```

```python
NUM_IMAGES = len(os.listdir(DATASET_PATH)) - 1
face_arr = np.zeros((62500, NUM_IMAGES))
```

```python
face_arr[:, 0] = np.array(img.getdata(), dtype=np.uint8)
```

```python
for i, img_dir in enumerate(os.listdir(DATASET_PATH)):
    if img_dir[0] == '.':
        continue
    face_arr[:, i - 1] = load_img(f'{DATASET_PATH}/{img_dir}/{img_dir}_0001.jpg')
```

```python
img = arr_to_img(np.mean(face_arr, axis=1))
```

```python
display(img)
```

```python
display(arr_to_img(np.min(face_arr, axis=1)))
```

```python
display(arr_to_img(np.max(face_arr, axis=1)))
```

```python
display(arr_to_img(np.median(face_arr, axis=1)))
```
